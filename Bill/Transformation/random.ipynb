```python
import requests
from bs4 import BeautifulSoup

# Base URL (without pagination)
BASE_URL = "https://finance.yahoo.com/research-hub/screener/sec-ind_ind-largest-equities_software-infrastructure/"

# HTTP headers to mimic a real browser
HEADERS = {"User-Agent": "Mozilla/5.0"}

# Pagination start parameter
start = 0
count = 25  # Number of results per page

while True:
    # Construct the paginated URL
    url = f"{BASE_URL}?start={start}&count={count}"
    response = requests.get(url, headers=HEADERS)

    # Stop if the request fails
    if response.status_code != 200:
        print(f"Failed to retrieve page {start//count + 1}. Status code: {response.status_code}")
        break

    # Parse HTML
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all <span> tags with class "symbol yf-1m808gl"
    symbols = soup.find_all("span", class_="symbol yf-1m808gl")

    # If no symbols are found, we've reached the last page
    if not symbols:
        print("No more data found. Stopping.")
        break

    # Extract and print the text content
    for symbol in symbols:
        print(symbol.get_text(strip=True))

    # Move to the next page
    start += count
```
