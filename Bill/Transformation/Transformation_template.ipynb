# **Cleaning code**

*Please be careful when running cells* - the toggled cells run as one so be EXTRA careful with this as we don't want to re-run the final cell in each toggle (this sends the table to BQ and will overwrite what we already have there). This isn't an issue right now but might be in the future!!

# what this code does:

**list of tables on gbq**    ----(download from gbq)---->     **list of data frames**
   ---(cleaning process)-->      **list of clean data frames**     -----(re-upload to gbq)---->  ** all those clean tables uploaded onto gbq**

# choose tables to change

```python
dataset="SP500_fin_london" #@param

tables_to_change=["fin_hist_highlow_all_SP500_thresh90"
  , "fin_hist_dividends_all_SP500_thresh90"
  , "fin_hist_mid_value_all_SP500_thresh90"
  , "fin_hist_stock_value_all_SP500_thresh90"
  , "fin_hist_volume_all_SP500_thresh90"
  , "fin_hist_splits_all_SP500_thresh90"
  , "fin_hist_total_value_all_SP500_thresh90"]
```

```python
#@title (all iterations)
# Query BigQuery to get the Before Ticker table for ESG data:

list_of_tables=['']*len(tables_to_change)

for i in tqdm(range(0,len(tables_to_change))):

  query_ = f"SELECT * FROM `lw-esg-project.{dataset}.{tables_to_change[i]}`"

  list_of_tables[i] = pd_gbq.read_gbq(query_, project_id = 'lw-esg-project')
```

```python
for i in list_of_tables:
  print(i.shape)
```

# clean data


```python
#@title (single iteration)
# list_of_tables[0].rename(columns={'Date':'date_date'},inplace=True)
```

```python
#@title (all iterations
for i in range(0,len(list_of_tables)):
    list_of_tables[i].rename(columns={'Date':'date_date'},inplace=True)
```

```python
for i in list_of_tables:
  print(i.shape)
```

```python
#@title (single iteration)
# Sending this table back to BQ and replacing the original table:
# pd_gbq.to_gbq(list_of_tables[0],dataset+'.'+tables_to_change[0] , project_id = 'lw-esg-project', if_exists = 'replace')
```

```python
#@title (all iterations
# Sending this table back to BQ and replacing the original table:
for i in range(0,len(list_of_tables)):
    pd_gbq.to_gbq(list_of_tables[i],dataset+'.'+tables_to_change[i] , project_id = 'lw-esg-project', if_exists = 'replace')
    print(tables_to_change[i],'updated')
```
