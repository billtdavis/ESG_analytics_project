```python
# CLEANING BAD TICKERS (they have gaps of more than a year so we're removing them)

# List of objects to check for in the columns
bad_tickers = ['AON', 'APA', 'CI', 'FOXA', 'IR', 'STX', 'ZION', 'BLK', 'LH', 'SPGI', 'USB', 'LB']

# Iterate through each column and check if the column name contains any of the bad tickers as PART of their name
columns_to_drop = []
for col in clean_df_before.columns:
    for obj in bad_tickers:
        # EXPLANATION OF THIS LOOP AT BOTTOM OF CELL:
        if obj in col.split('esgScore')[-1] and len(col.split('esgScore')[-1])==len(obj) and col.startswith('esgScore'):
            columns_to_drop.append(col)
        if obj in col.split('governanceScore')[-1] and len(col.split('governanceScore')[-1])==len(obj) and col.startswith('governanceScore'):
            columns_to_drop.append(col)
        if obj in col.split('socialScore')[-1] and len(col.split('socialScore')[-1])==len(obj) and col.startswith('socialScore'):
            columns_to_drop.append(col)
        if obj in col.split('environmentScore')[-1] and len(col.split('environmentScore')[-1])==len(obj) and col.startswith('environmentScore'):
            columns_to_drop.append(col)
            break

print(columns_to_drop) # Columns that are dropped
print(len(columns_to_drop)) # Number of columns that are dropped (48, 4x 12 tickers)

# Drop the identified columns
clean_df_before = clean_df_before.drop(columns=columns_to_drop)

clean_df_before.shape # Checking the shape of the table matches what we expected (1293 columns)




# EXPLANATION OF FOR LOOP (down here for clarity of the code above)
# CODE for obj in bad_tickers:
# CODE    if obj in col.split('esgScore')[-1] and len(col.split('esgScore')[-1])==len(obj) and col.startswith('esgScore'):
          # 'if obj in col.split('esgScore')[-1]' -> checks if the current obj (the ticker in the bad_tickers list) is found in the column name
              # 'col.split('esgScore')' -> splits the column name into parts using 'esgScore' as the delimiter
              # '[-1]' -> selects the last part of the split column name (the ticker) to check it
          # 'and len(col.split('esgScore')[-1])==len(obj)' -> ensures that the length of the isolated ticker in the column matches the length of the obj from bad_tickers (so we match the WHOLE ticker, not part of it)
          # 'and col.startswith('esgScore'):' -> verifies that the column name starts with 'esgScore'
# CODE       columns_to_drop.append(col)
              # If all 3 conditions are True, the column name will be added to the columns_to_drop list
# CODE       break
             # exits the inner loop as soon as a match is found, as there should only be one column with this name, there is no need to check the rest
```

```python
#@title (Melting all the tables:)
dataset="sp500_esg_breakdown_london" #@param

tables_to_change = ["clean_ESG_ONLY_before_ticker_SP500"
                  , "clean_ENV_ONLY_before_ticker_SP500"
                  , "clean_SOC_ONLY_before_ticker_SP500"
                  , "clean_GOV_ONLY_before_ticker_SP500"]
```

```python
import pandas_gbq as pd_gbq
from tqdm import tqdm
```

```python
# STEP 1: Query BQ to get all the tables into colab:
list_of_tables=['']*len(tables_to_change)

for i in tqdm(range(0,len(tables_to_change))):

  query_ = f"SELECT * FROM `lw-esg-project.{dataset}.{tables_to_change[i]}`"

  list_of_tables[i] = pd_gbq.read_gbq(query_, project_id = 'lw-esg-project')
```

```python
# STEP 2: Renaming columns to just tickers across all split tables (ESG_Score):
new_columns = []
for col in list_of_tables[0].columns:
  col.split('esgScore')[-1]
  new_columns.append(col.split('esgScore')[-1])
list_of_tables[0].columns = new_columns

list_of_tables[0].head()
```

```python
# STEP 2: Renaming columns to just tickers across all split tables (E_Score):
new_columns = []
for col in list_of_tables[1].columns:
  col.split('environmentScore')[-1]
  new_columns.append(col.split('environmentScore')[-1])
list_of_tables[1].columns = new_columns

list_of_tables[1].head()
```

```python
# STEP 2: Renaming columns to just tickers across all split tables (S_Score):
new_columns = []
for col in list_of_tables[2].columns:
  col.split('socialScore')[-1]
  new_columns.append(col.split('socialScore')[-1])
list_of_tables[2].columns = new_columns

list_of_tables[2].head()
```

```python
# STEP 2: Renaming columns to just tickers across all split tables (G_Score):
new_columns = []
for col in list_of_tables[3].columns:
  col.split('governanceScore')[-1]
  new_columns.append(col.split('governanceScore')[-1])
list_of_tables[3].columns = new_columns

list_of_tables[3].head()
```

```python
# STEP 3: Melt each split table to make tables that Looker can work nicely with:

list_of_tables[0] = pd.melt(frame = list_of_tables[0], id_vars = 'date_date', var_name = 'Ticker', value_name = 'ESG_Score')

list_of_tables[0].head()
```

```python
# STEP 3: Melt each split table to make tables that Looker can work nicely with:

list_of_tables[1] = pd.melt(frame = list_of_tables[1], id_vars = 'date_date', var_name = 'Ticker', value_name = 'E_Score')

list_of_tables[1].head()
```

```python
# STEP 3: Melt each split table to make tables that Looker can work nicely with:
list_of_tables[2] = pd.melt(frame = list_of_tables[2], id_vars = 'date_date', var_name = 'Ticker', value_name = 'S_Score')

list_of_tables[2].head()
```

```python
# STEP 3: Melt each split table to make tables that Looker can work nicely with:

list_of_tables[3] = pd.melt(frame = list_of_tables[3], id_vars = 'date_date', var_name = 'Ticker', value_name = 'G_Score')

list_of_tables[3].head()
```

```python
# STEP 4: Creating primary key (which is a join of date_date and Ticker):
list_of_tables[0]["key"] = list_of_tables[0]["date_date"].astype(str) + "-" + list_of_tables[0]["Ticker"]

list_of_tables[0].head()
```

```python
# STEP 4: Creating primary key (which is a join of date_date and Ticker):
list_of_tables[1]["key"] = list_of_tables[1]["date_date"].astype(str) + "-" + list_of_tables[1]["Ticker"]

list_of_tables[1].head()
```

```python
# STEP 4: Creating primary key (which is a join of date_date and Ticker):
list_of_tables[2]["key"] = list_of_tables[2]["date_date"].astype(str) + "-" + list_of_tables[2]["Ticker"]

list_of_tables[2].head()
```

```python
# STEP 4: Creating primary key (which is a join of date_date and Ticker):
list_of_tables[3]["key"] = list_of_tables[3]["date_date"].astype(str) + "-" + list_of_tables[3]["Ticker"]

list_of_tables[3].head()
```

```python
# STEP 5: Joining the melted ESG table to the melted E_Score table using their indexes from the list_of_tables variable:

melted_joined = list_of_tables[0].merge(list_of_tables[1], how = 'outer', on = 'key', suffixes = ('', 'extra'))
# The suffixes argument means that any columns names that are duplicates will have extra added to the end of their name (ONLY for the right
# hand side columns, the left ones remain the same)

melted_joined.drop(columns = melted_joined.filter(regex = 'extra$').columns, inplace = True)
# Adding these suffixes means we can then filter by them and drop the duplicate columns

melted_joined.head()
```

```python
# STEP 6: Joining all the melted tables together into one table (melted_joined):
melted_joined = melted_joined.merge(list_of_tables[2], how = 'outer', on = 'key', suffixes = ('', 'extra'))
melted_joined.drop(columns = melted_joined.filter(regex = 'extra$').columns, inplace = True)

melted_joined.head()
```

```python
# STEP 6: Joining all the melted tables together into one table (melted_joined):
melted_joined = melted_joined.merge(list_of_tables[3], how = 'outer', on = 'key', suffixes = ('', 'extra'))
melted_joined.drop(columns = melted_joined.filter(regex = 'extra$').columns, inplace = True)

melted_joined.head()
```

```python
# STEP 7: Moving the key column to the far left:
cols = ['key'] + [col for col in melted_joined.columns if col != 'key']
melted_joined = melted_joined[cols]

melted_joined.head()
```

```python
# STEP 8: Sending this melted table back to BQ: THIS IS ALL OF THE BEFORE TABLES MELTED:
pandas_gbq.to_gbq(melted_joined, 'sp500_esg_breakdown_london.melted_joined_before_ticker_SP500', project_id = 'lw-esg-project', if_exists = 'replace')
```

```python
# CHECKING FOR NULLS IN THE AFTER ESG DF:

# Check for null values in the DataFrame
null_counts_after = after_df.isnull().sum()
print(null_counts_after)

# Check if any column has null values
has_nulls_after = after_df.isnull().values.any()
print(f"\nDataFrame has null values: {has_nulls_after}")

# Filter to show only columns with nulls
null_columns_after = null_counts_after[null_counts_after > 20]

# Print the result
print("Columns with null values:")
print(null_columns_after)
print("Number of null columns: ", len(null_columns_after))
```

```python
# CLEANING BAD TICKERS (they have gaps of more than a year so we're removing them)

# List of objects to check for in the columns
bad_tickers = ['AON', 'APA', 'CI', 'FOXA', 'IR', 'STX', 'ZION', 'BLK', 'LH', 'SPGI', 'USB', 'LB', 'TPR', 'BKNG', 'BRK-B', 'LUMN']

# Iterate through each column and check if the column name contains any of the bad tickers as PART of their name
columns_to_drop = []
for col in clean_df_after.columns:
    for obj in bad_tickers:
        # EXPLANATION OF THIS LOOP AT BOTTOM OF CELL:
        if obj in col.split('esgScore')[-1] and len(col.split('esgScore')[-1])==len(obj) and col.startswith('esgScore'):
            columns_to_drop.append(col)
        if obj in col.split('governanceScore')[-1] and len(col.split('governanceScore')[-1])==len(obj) and col.startswith('governanceScore'):
            columns_to_drop.append(col)
        if obj in col.split('socialScore')[-1] and len(col.split('socialScore')[-1])==len(obj) and col.startswith('socialScore'):
            columns_to_drop.append(col)
        if obj in col.split('environmentScore')[-1] and len(col.split('environmentScore')[-1])==len(obj) and col.startswith('environmentScore'):
            columns_to_drop.append(col)
            break

print(columns_to_drop) # Columns that are dropped
print(len(columns_to_drop)) # Number of columns that are dropped (48, 4x 12 tickers)

# Drop the identified columns
clean_df_after = clean_df_after.drop(columns=columns_to_drop)

clean_df_after.shape # Checking the shape of the table matches what we expected (1293 columns)




# EXPLANATION OF FOR LOOP (down here for clarity of the code above)
# CODE for obj in bad_tickers:
# CODE    if obj in col.split('esgScore')[-1] and len(col.split('esgScore')[-1])==len(obj) and col.startswith('esgScore'):
          # 'if obj in col.split('esgScore')[-1]' -> checks if the current obj (the ticker in the bad_tickers list) is found in the column name
              # 'col.split('esgScore')' -> splits the column name into parts using 'esgScore' as the delimiter
              # '[-1]' -> selects the last part of the split column name (the ticker) to check it
          # 'and len(col.split('esgScore')[-1])==len(obj)' -> ensures that the length of the isolated ticker in the column matches the length of the obj from bad_tickers (so we match the WHOLE ticker, not part of it)
          # 'and col.startswith('esgScore'):' -> verifies that the column name starts with 'esgScore'
# CODE       columns_to_drop.append(col)
              # If all 3 conditions are True, the column name will be added to the columns_to_drop list
# CODE       break
             # exits the inner loop as soon as a match is found, as there should only be one column with this name, there is no need to check the rest
```
